```python
import os
import gzip
import shutil

# Input directory containing the .json.gz files
input_dir = '/content/drive/MyDrive/VR Project 2/Data/abo-listings/listings/metadata/'

# Output directory where the unzipped .json files will be saved
output_dir = '/content/drive/MyDrive/VR Project 2/Data/abo-listings/listings/unzipped-metadata/'

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Loop through all files in the input directory
for file_name in os.listdir(input_dir):
    if file_name.endswith('.json.gz'):
        # Define the full path for the input .json.gz file
        gz_file_path = os.path.join(input_dir, file_name)

        # Define the output file path (same name, but without .gz extension)
        output_file_path = os.path.join(output_dir, file_name.replace('.gz', ''))

        # Open the .gz file and the output file
        with gzip.open(gz_file_path, 'rb') as f_in:
            with open(output_file_path, 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)

        print(f"Unzipped: {file_name} -> {output_file_path}")

print("All files have been unzipped!")
```

```python
import os
import pandas as pd
import json

# Directory containing the unzipped .json files
unzipped_dir = '/content/drive/MyDrive/VR Project 2/Data/abo-listings/listings/unzipped-metadata/'

# Output file path where the combined JSON file will be saved
combined_json_file = '/content/drive/MyDrive/VR Project 2/Data/abo-listings/listings/combined_listings.json'

# List to hold data from all files
combined_data = []

# Loop through all the files in the unzipped directory
for file_name in os.listdir(unzipped_dir):
    if file_name.endswith('.json'):
        file_path = os.path.join(unzipped_dir, file_name)

        try:
            # Try loading the file as JSON lines (if each line is a separate JSON object)
            df = pd.read_json(file_path, lines=True)  # Specify lines=True for JSON Lines format
            combined_data.append(df)
        except ValueError as e:
            print(f"ValueError: Trailing data or invalid format in file: {file_name}")
            print(f"Error: {e}")
        except json.JSONDecodeError as e:
            print(f"JSONDecodeError in file: {file_name}")
            print(f"Error: {e}")

# Concatenate all DataFrames into one
if combined_data:
    combined_df = pd.concat(combined_data, ignore_index=True)

    # Save the combined DataFrame to a single JSON file
    combined_df.to_json(combined_json_file, orient='records', lines=True)
    print(f"All JSON files have been combined into: {combined_json_file}")
else:
    print("No valid JSON data to combine.")
```

```python
import pandas as pd
import json

# === Paths ===
json_path = '/content/drive/MyDrive/VR Project 2/Data/abo-listings/listings/combined_listings.json'
csv_path = '/content/drive/MyDrive/VR Project 2/Data/abo-images-small/images/metadata/images.csv'

# === Load image metadata ===
image_df = pd.read_csv(csv_path)

# === Load JSON product records ===
json_records = []
with open(json_path, 'r') as f:
    for line in f:
        try:
            record = json.loads(line.strip())
            json_records.append(record)
        except json.JSONDecodeError as e:
            print(f"Skipping invalid JSON line: {e}")

# === Build priority-aware image_id → product mapping ===
main_image_to_product = {}
other_image_to_product = {}

for record in json_records:
    main_id = record.get('main_image_id')
    if main_id and main_id not in main_image_to_product:
        main_image_to_product[main_id] = record

    other_ids = record.get('other_image_id', [])
    if isinstance(other_ids, list):
        for other_id in other_ids:
            if (
                other_id not in main_image_to_product and
                other_id not in other_image_to_product
            ):
                other_image_to_product[other_id] = record

# Merge mappings with priority: main > other
image_to_product = {**other_image_to_product, **main_image_to_product}

# Convert to DataFrame
product_link_df = pd.DataFrame([
    {'image_id': img_id, 'product_data': metadata}
    for img_id, metadata in image_to_product.items()
])

# Flatten product_data only up to 1 level
flattened_product_df = pd.json_normalize(product_link_df['product_data'], max_level=1)

# Merge flattened fields with image_id
product_link_df = pd.concat([product_link_df[['image_id']], flattened_product_df], axis=1)

# Final merge with image metadata
merged_df = pd.merge(image_df, product_link_df, on='image_id', how='left')

# Save result
output_csv = '/content/drive/MyDrive/VR Project 2/Data/combined_flattened_level1.csv'
merged_df.to_csv(output_csv, index=False)

print("✅ CSV saved with image metadata and product info flattened up to one level.")
```

```python
import pandas as pd
import ast

# Load your merged data
df = pd.read_csv('/content/drive/MyDrive/VR Project 2/Data/combined_flattened_level1.csv')

# Define which columns contain language-tagged fields
multi_value_cols = ['item_name', 'bullet_point', 'item_keywords']  # return list of strings
single_value_cols = ['brand', 'color']  # return first English value

def extract_english_values(cell, return_list=False):
    if pd.isna(cell):
        return None
    try:
        data = ast.literal_eval(cell) if isinstance(cell, str) else cell
        if isinstance(data, list):
            en_values = [entry.get('value') for entry in data
                         if isinstance(entry, dict) and entry.get('language_tag', '').startswith('en')]
            if not en_values:
                return None
            return en_values if return_list else en_values[0]
    except Exception:
        pass
    return None

# Clean the DataFrame
cleaned_df = df.copy()

# Process multi-value columns (keep all English values)
for col in multi_value_cols:
    if col in cleaned_df.columns:
        cleaned_df[col] = cleaned_df[col].apply(lambda x: extract_english_values(x, return_list=True))

# Process single-value columns (keep only first English value)
for col in single_value_cols:
    if col in cleaned_df.columns:
        cleaned_df[col] = cleaned_df[col].apply(lambda x: extract_english_values(x, return_list=False))

# Optional: save cleaned file
output_path = '/content/drive/MyDrive/VR Project 2/Data/final_english_cleaned.csv'
cleaned_df.to_csv(output_path, index=False)

print("✅ Cleaned DataFrame saved with all English-language values (including multiple entries where needed).")
```
